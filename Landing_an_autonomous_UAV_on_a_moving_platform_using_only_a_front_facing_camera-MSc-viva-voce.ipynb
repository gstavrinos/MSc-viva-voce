<h2>Global Planning</h2>
<hr>
<br><br>

</center>



<h3>


Input:
<ol>
    <li> Current robot position.
    <li> Latest known position of the moving platform.
    <li> Latest set of estimated velocities.
</ol>

<br>
<hr>
Velocity estimation noise smoothing by removing values that are not close to the mean of each set.
<hr>

<br>
Two different strategies to choose the goal point:
<ol>
    <li> First point that can be reached based on kinematic limitations.
    <li> Point that requires the lowest velocity.
</ol>

<h1><span style="color:#8e1731">Chapter 12</span></h1>
<h2>Local planning</h2>
<hr>

<br><br>

</center>

<h3>
Input:
<ol>
    <li> Global plan.
    <li> Current robot position.
    <li> Current estimated platform position.
</ol>
<hr>
<center>
Linear velocity:
<hr>
\begin{equation}
\begin{cases} v_0 = 2 \times \sqrt{(x_r - x_h)^2 + (y_r - y_h)^2} \times \cos{(\tan^{-1}{(\frac{y_h - y_r}{x_h - x_r})} - yaw_r)} \\
v_x = v_0, v_0\in[v_{min},v_{max}] \\
v_x = v_{min}, v_0\in(-\inf,v_{min}) \\ 
v_x = v_{max}, v_0\in(v_{max},+\inf) \end{cases}
\end{equation}
<hr>
\begin{equation}
\begin{cases} v_0 = 2 \times \sqrt{(x_r - x_h)^2 + (y_r - y_h)^2} \times \sin{(\tan^{-1}{(\frac{y_h - y_r}{x_h - x_r})} - yaw_r)} \\
v_y = v_0, v_0\in[v_{min},v_{max}] \\
v_y = v_{min}, v_0\in(-\inf,v_{min}) \\ 
v_y = v_{max}, v_0\in(v_{max},+\inf) \end{cases}
\end{equation}
<hr>
\begin{equation}
v_z = 2 \times (z_h - (z_r - 0.1))
\end{equation}
<hr>
Angular Velocity:
<hr>
\begin{equation}
v_{yaw} = (\tan^{-1}{(\frac{y_h - y_r}{x_h - x_r})} - yaw_r) \times 10
\end{equation}
<hr>
</center>

Send the landing command when:
<ol>
    <li> The two dimensional distance between the robot and the moving platform on the x and y axes is less than 20cm.
    <li> The distance (difference) in the z axis between the robot and the moving platform is less than 20cm.
    <li> The robot is above the moving platform.
</ol>
    
<br>
Note: When a landing command is sent, always the emergency command is used.

</h3>
